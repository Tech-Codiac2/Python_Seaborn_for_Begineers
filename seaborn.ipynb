```python
import pandas as pd
```

```python
import seaborn as sns
sns.set(color_codes=True)
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
auto=pd.read_csv('C:\\Users\\91750\\OneDrive\\Documents\\Automobile.csv')
```

```python
auto.head()
```

```python
sns.displot(auto['normalized_losses'],rug=True)
plt.show()
```

```python
sns.jointplot(auto['engine_size'],auto['horsepower'])
```

```python
sns.jointplot(auto['engine_size'],auto['horsepower'], kind='hex') ##can use kind='reg' for linear regression, 'kde', for kde
plt.show()
```

```python
#multivariate plots

sns.pairplot(auto[['normalized_losses','engine_size']])
```

```python
sns.stripplot(auto['fuel_type'],auto['horsepower'] ,jitter=True)
```

```python
sns.swarmplot(auto['fuel_type'],auto['horsepower'] )
```

```python
sns.boxplot(auto['fuel_type'],auto['horsepower'] )
```

```python
sns.boxplot(auto['body_style'],auto['horsepower'],hue=auto['engine_location']  )
```

```python
sns.countplot(auto['body_style'])
```

```python
sns.pointplot(auto['body_style'],auto['horsepower'])  
```
